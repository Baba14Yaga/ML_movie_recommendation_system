```python
def convert(obj): #converts the string into list 
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj): #extracting first 3 cast from movies
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
            
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj): #extracting director name from crew members
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
def convert1(obj): #converts the string into list 
    if obj==' ':
        return ' '
    obj=ast.literal_eval(obj) 
    return obj['name']
```

```python
convert1(movies.iloc[0].belongs_to_collection)
```

```python
movies['belongs_to_collection']=movies['belongs_to_collection'].apply(convert1)
```

```python
movies['tags']=movies['overview'] +movies['genres'] + movies ['keywords']+movies['cast'] +movies['crew'] # storing all tags in single column
```

```python
new_df =movies[['id','title','tags','tagline','belongs_to_collection']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
#converts tags into strings
```

```python
new_df['tags']=new_df['tags']+new_df['tagline']+new_df['tagline']
```

```python
new_df['tags']= new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
new_df[new_df['title']=='Spectre'].index[0]
```

```python
#as you can see that there are many words in above list with same meaning but with -ed,-s,etc for eg action and actions
#for better prediction these words must be combined in single dimension
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)  
```

```python
new_df['tags']=new_df['tags'].apply(stem)
new_df.reset_index(inplace=True)
```

```python
#performing count vectorization 
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
vector1=cv.fit_transform(new_df['tags']).toarray()
cv.get_feature_names()
from sklearn.metrics.pairwise import cosine_similarity
similarity1=cosine_similarity(vector1) #calculating the cosine distances in multidimensional array
```

```python
def recommend1(movie):
    movie_index =new_df[new_df['title']==movie].index[0]
    distances=similarity1[movie_index]
    movies_list=sorted (list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:21]
    #above line will pair the index with cosine values so that it do not gets lost 
    L=[]
    for i in movies_list:
        L.append(i[0])
    return L
```

```python
new_df['BagOfWords']=new_df['title'].apply(recommend1)
```

```python
#performing TF-IDF on tags from new_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(max_features=6000,stop_words='english')
```

```python
vectorizer.fit_transform(new_df['tags'])
```

```python
vectors=vectorizer.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
vectorizer.get_feature_names()
```

```python
#now we will calculate the cosine distance(angle between vectors) in higher dimensions as it is more reliable then eucldian distance
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors) #calculating the cosine distances in multidimensional array
```

```python
similarity
```

```python
def recommend(movie):
    # function to store recommended movies index so that it could be later used for showing list  
    movie_index =new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted (list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:21]
    #above line will pair the index with cosine values so that it do not gets lost 
    recommendation=[]
    for i in movies_list:
        recommendation.append(i[0])
    return recommendation
```

```python
recommend('Fast Five')
```

```python
new_df['TfIdf']=new_df['title'].apply(recommend) #applying recommend to all the rows of movies
```

```python
new_df=new_df[['id','title','TfIdf','BagOfWords']]
```
